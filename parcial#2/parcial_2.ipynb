```python
%%writefile pages/1_Modulacion_AM.py
import streamlit as st
import numpy as np
from scipy.fft import fft, ifft, fftshift, fftfreq
from scipy.io import wavfile
import matplotlib.pyplot as plt
import yt_dlp
import os
import subprocess

st.set_page_config(page_title="Punto 1: Modulaci√≥n AM", page_icon="üìª", layout="wide")

st.markdown("""
<div style="border-bottom: 2px solid #1E90FF; margin-bottom: 20px;">
    <h2 style="color: #1E90FF;">üìª Punto 1: Modulaci√≥n y Demodulaci√≥n (DSB-SC)</h2>
</div>
""", unsafe_allow_html=True)

# --- SIDEBAR ---
st.sidebar.header("üéõÔ∏è Par√°metros de Se√±al")
url = st.sidebar.text_input("URL YouTube", "https://www.youtube.com/watch?v=5qap5aO4i9A")
start_s = st.sidebar.number_input("Segundo Inicio", value=30)
fc = st.sidebar.slider("Frecuencia Portadora (Hz)", 1000, 15000, 5000)
cutoff = st.sidebar.slider("Frecuencia Corte Filtro (Hz)", 1000, 8000, 4000)

# --- FUNCIONES ---
@st.cache_data
def get_audio(url, start, dur=5):
    """Descarga y procesa audio de YT"""
    out = "audio.wav"
    opts = {'format':'bestaudio/best', 'outtmpl':'temp', 'quiet':True}
    try:
        if os.path.exists(out): os.remove(out)
        with yt_dlp.YoutubeDL(opts) as ydl:
            ydl.download([url])
        # Convertir a WAV mono 44.1kHz con ffmpeg
        subprocess.run(
            ['ffmpeg', '-i', 'temp', '-ss', str(start), '-t', str(dur),
             '-ac', '1', '-ar', '44100', out, '-y'],
            stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL
        )
        fs, data = wavfile.read(out)
        # Normalizar
        if data.dtype != np.float32:
            data = data.astype(np.float32) / np.max(np.abs(data))
        return fs, data
    except Exception as e:
        return None, None

def plot_tf(t, sig, fs, title, color):
    """Grafica Tiempo y Frecuencia"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 3.5))
    # Tiempo
    ax1.plot(t, sig, color=color, lw=0.8)
    ax1.set_title(f"Tiempo: {title}")
    ax1.grid(alpha=0.3)
    # Frecuencia
    N = len(sig)
    Y = fftshift(fft(sig))
    f = fftshift(fftfreq(N, 1/fs))
    ax2.plot(f, np.abs(Y)/N, color=color, lw=0.8)
    ax2.set_title(f"Espectro: {title}")
    ax2.set_xlim(-fs/2, fs/2)
    ax2.grid(alpha=0.3)
    st.pyplot(fig)

# --- APP ---
if st.sidebar.button("‚ñ∂Ô∏è Iniciar Simulaci√≥n"):
    with st.spinner("Descargando audio..."):
        fs, m_t = get_audio(url, start_s)

    if m_t is not None:
        N = len(m_t)
        t = np.linspace(0, 5, N)

        st.info("1. Se√±al Mensaje Original m(t)")
        st.audio(m_t, sample_rate=fs)
        plot_tf(t, m_t, fs, "Mensaje", "#1E90FF")

        # Modulaci√≥n
        c_t = np.cos(2*np.pi*fc*t)
        y_t = m_t * c_t
        st.info("2. Se√±al Modulada (DSB-SC)")
        st.audio(y_t, sample_rate=fs)
        plot_tf(t, y_t, fs, "Modulada", "#28a745")

        # Demodulaci√≥n (Mezcla)
        y_mix = y_t * c_t
        st.info("3. Se√±al Mezclada (Antes del Filtro)")
        plot_tf(t, y_mix, fs, "Mezclador", "#dc3545")

        # Filtro Ideal FFT
        Y_f = fft(y_mix)
        freqs = fftfreq(N, 1/fs)
        mask = np.abs(freqs) < cutoff
        m_rec = np.real(ifft(Y_f * mask)) * 2 # x2 ganancia

        st.success("4. Se√±al Recuperada (Filtrado Ideal)")
        st.audio(m_rec, sample_rate=fs)
        plot_tf(t, m_rec, fs, "Recuperada", "#1E90FF")
    else:
        st.error("Error al descargar audio.")
else:
    st.info("Presiona el bot√≥n en la barra lateral para comenzar.")
```

```python
%%writefile pages/2_Sistema_Masa_Resorte.py
import streamlit as st
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

# 1. Configuraci√≥n
st.set_page_config(page_title="Punto 2: Sistemas 2do Orden", page_icon="‚öôÔ∏è", layout="wide")

# 2. Estilos CSS Personalizados
st.markdown("""
    <style>
    /* Header con degradado √Åmbar/Naranja */
    .main-header {
        background: linear-gradient(135deg, #FFC107 0%, #FF8C00 100%);
        padding: 20px;
        border-radius: 15px;
        color: white;
        box-shadow: 0 4px 15px rgba(0,0,0,0.1);
        margin-bottom: 25px;
        text-align: center;
    }

    /* Tarjetas de M√©tricas (Analog√≠as) */
    .metric-container {
        background-color: #f8f9fa;
        border-radius: 10px;
        padding: 15px;
        border-left: 5px solid #FFC107;
        box-shadow: 0 2px 5px rgba(0,0,0,0.05);
        height: 100%;
    }
    .metric-title {
        font-weight: bold;
        color: #555;
        font-size: 1.1rem;
        margin-bottom: 10px;
        border-bottom: 1px solid #ddd;
        padding-bottom: 5px;
    }
    .metric-value {
        font-size: 0.95rem;
        color: #333;
        margin: 5px 0;
    }
    .metric-label {
        color: #777;
        font-size: 0.85rem;
    }

    /* Ajuste Modo Oscuro */
    @media (prefers-color-scheme: dark) {
        .metric-container {
            background-color: #262730;
            border-left: 5px solid #FFC107;
        }
        .metric-title { color: #ddd; border-color: #444; }
        .metric-value { color: #fff; }
        .metric-label { color: #aaa; }
    }
    </style>
""", unsafe_allow_html=True)

# 3. Encabezado
st.markdown("""
    <div class="main-header">
        <h2 style="margin:0; color:white;">‚öôÔ∏è Punto 2: Din√°mica de Sistemas de 2do Orden</h2>
        <p style="margin:5px 0 0 0; opacity:0.9;">Analog√≠a Electromec√°nica y An√°lisis de Estabilidad</p>
    </div>
""", unsafe_allow_html=True)

# --- SIDEBAR ---
st.sidebar.markdown("### üéõÔ∏è Configuraci√≥n del Sistema")
st.sidebar.info("Ajusta los par√°metros f√≠sicos para ver la respuesta en tiempo real.")

caso = st.sidebar.selectbox("Tipo de Respuesta", ["Subamortiguado (Oscila)", "Cr√≠tico", "Sobreamortiguado"])

# L√≥gica de Sliders condicionales
if "Sub" in caso:
    zeta = st.sidebar.slider("Zeta (Œ∂) - Amortiguamiento", 0.05, 0.95, 0.3)
elif "Cr√≠t" in caso:
    zeta = 1.0
    st.sidebar.warning("üîí Zeta fijo en 1.0 para caso Cr√≠tico")
else:
    zeta = st.sidebar.slider("Zeta (Œ∂) - Amortiguamiento", 1.05, 5.0, 1.5)

wn = st.sidebar.slider("Omega n (œân) - Frec. Natural", 1.0, 20.0, 5.0)
st.sidebar.markdown("---")
lazo = st.sidebar.radio("Arquitectura de Control", ["Lazo Abierto", "Lazo Cerrado"], horizontal=True)

# --- C√ÅLCULOS ---
# Mec√°nico (Normalizado m=1)
m = 1.0
k = (wn**2) * m
c_dump = 2 * zeta * wn * m # Rename c to c_dump to avoid conflict with st.columns variables

# El√©ctrico (Serie RLC, asumiendo C=1mF)
C_el = 0.001
L_el = 1 / (wn**2 * C_el)
R_el = 2 * zeta * wn * L_el

# Funciones de Transferencia
num = [wn**2]
den = [1, 2*zeta*wn, wn**2]

if lazo == "Lazo Cerrado":
    # G_cl = G / (1+G) -> Denominador cambia: s^2 + 2*zeta*wn*s + 2*wn^2
    den_sys = [den[0], den[1], den[2] + num[0]]
    sys = signal.TransferFunction(num, den_sys)
    wn_eff = np.sqrt(den_sys[2])
    z_eff = den_sys[1]/(2*wn_eff)
    tf_latex = f"G_{{CL}}(s) = \\frac{{{wn**2:.1f}}}{{s^2 + {den_sys[1]:.1f}s + {den_sys[2]:.1f}}}"
else:
    sys = signal.TransferFunction(num, den)
    wn_eff = wn
    z_eff = zeta
    tf_latex = f"G_{{OL}}(s) = \\frac{{{wn**2:.1f}}}{{s^2 + {den[1]:.1f}s + {den[2]:.1f}}}"

# --- DASHBOARD SUPERIOR (M√©tricas y Ecuaci√≥n) ---
col_eq, col_metrics = st.columns([1, 2])

with col_eq:
    st.markdown("##### Funci√≥n de Transferencia")
    st.latex(tf_latex)
    st.caption(f"Configuraci√≥n: **{lazo}**")

with col_metrics:
    c1, c2, c3 = st.columns(3)
    # Card 1: Mec√°nico
    c1.markdown(f"""
    <div class="metric-container">
        <div class="metric-title">üõ†Ô∏è Mec√°nico</div>
        <div class="metric-value">{k:.2f} <span class="metric-label">N/m (k)</span></div>
        <div class="metric-value">{c_dump:.2f} <span class="metric-label">N¬∑s/m (c)</span></div>
    </div>
    """, unsafe_allow_html=True)

    # Card 2: El√©ctrico
    c2.markdown(f"""
    <div class="metric-container" style="border-left-color: #0dcaf0;">
        <div class="metric-title">‚ö° El√©ctrico</div>
        <div class="metric-value">{L_el:.3f} <span class="metric-label">H (L)</span></div>
        <div class="metric-value">{R_el:.2f} <span class="metric-label">Œ© (R)</span></div>
    </div>
    """, unsafe_allow_html=True)

    # Card 3: Sistema Efectivo
    c3.markdown(f"""
    <div class="metric-container" style="border-left-color: #28a745;">
        <div class="metric-title">üìà Efectivo</div>
        <div class="metric-value">{wn_eff:.2f} <span class="metric-label">rad/s (œân)</span></div>
        <div class="metric-value">{z_eff:.3f} <span class="metric-label">Zeta (Œ∂)</span></div>
    </div>
    """, unsafe_allow_html=True)

st.write("") # Espacio

# --- PESTA√ëAS DE AN√ÅLISIS ---
tab1, tab2, tab3 = st.tabs(["üìä Respuesta al Escal√≥n", "„Ä∞Ô∏è Diagrama de Bode", "‚úñÔ∏è Mapa de Polos y Ceros"])

# Configuraci√≥n global de plots
plt.rcParams['figure.autolayout'] = True
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False

with tab1:
    t, y = signal.step(sys)
    fig, ax = plt.subplots(figsize=(10, 4))

    # Plot principal
    ax.plot(t, y, lw=2.5, color='#FF8C00', label='Salida y(t)')
    ax.axhline(y[-1], color='gray', linestyle='--', alpha=0.5, label='Estado Estacionario')
    ax.axhline(1.0, color='green', linestyle=':', alpha=0.4, label='Referencia (Step)')

    # Estilizado
    ax.set_title(f"Respuesta Transitoria ({caso})", fontsize=12, fontweight='bold')
    ax.set_xlabel("Tiempo (segundos)")
    ax.set_ylabel("Amplitud")
    ax.grid(True, alpha=0.2, linestyle='--')
    ax.legend(loc='lower right', frameon=False)
    ax.set_facecolor('#fafafa') # Fondo muy suave

    st.pyplot(fig)

    # Datos clave del transitorio
    peak = np.max(y)
    overshoot = (peak - y[-1])/y[-1] * 100 if peak > y[-1] else 0
    st.caption(f"**An√°lisis R√°pido:** Sobrepaso (Overshoot): {overshoot:.1f}% | Valor Final: {y[-1]:.3f}")

with tab2:
    w, mag, phase = signal.bode(sys)
    fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 5))

    # Magnitud
    ax1.semilogx(w, mag, color='#6f42c1', lw=2)
    ax1.set_ylabel("Magnitud (dB)", color='#6f42c1')
    ax1.grid(True, which="both", alpha=0.2)
    ax1.set_title("Respuesta en Frecuencia (Bode)", fontsize=12, fontweight='bold')

    # Fase
    ax2.semilogx(w, phase, color='#20c997', lw=2)
    ax2.set_ylabel("Fase (grados)", color='#20c997')
    ax2.set_xlabel("Frecuencia (rad/s)")
    ax2.grid(True, which="both", alpha=0.2)

    st.pyplot(fig)

with tab3:
    fig, ax = plt.subplots(figsize=(6, 5))

    # Ejes
    ax.axhline(0, color='black', lw=1)
    ax.axvline(0, color='black', lw=1)

    # Polos y Ceros
    p_real = np.real(sys.poles)
    p_imag = np.imag(sys.poles)

    ax.scatter(p_real, p_imag, marker='x', s=150, color='#dc3545', label=f'Polos ({len(p_real)})', zorder=5)

    if len(sys.zeros) > 0:
        ax.scatter(np.real(sys.zeros), np.imag(sys.zeros), marker='o', s=120, facecolors='none', edgecolors='#0d6efd', linewidth=2, label='Ceros')

    ax.grid(True, alpha=0.3, linestyle='--')
    ax.set_title("Mapa de Polos y Ceros (Plano S)", fontweight='bold')
    ax.set_xlabel("Eje Real (œÉ)")
    ax.set_ylabel("Eje Imaginario (jœâ)")
    ax.legend()

    # Zona de estabilidad visual
    if np.all(p_real < 0):
        st.success("‚úÖ El sistema es **ESTABLE** (Todos los polos en el semiplano izquierdo).")
    else:
        st.error("‚ö†Ô∏è El sistema es **INESTABLE**.")

    col_plot, col_info = st.columns([2, 1])
    with col_plot:
        st.pyplot(fig)
    with col_info:
        st.write("**Coordenadas de Polos:**")
        for p in sys.poles:
            st.code(f"{p.real:.2f} + {p.imag:.2f}j")
```
